```python
filenames = []
for file in os.listdir('t20s'):
    filenames.append(os.path.join('t20s',file))
```

```python
filenames[0:100]
```

```python
from yaml import safe_load
import pandas as pd
from tqdm import tqdm
import os

batch = []
batch_num = 0
match_id = 0

for file in tqdm(filenames, desc="Extracting"):

    try:
        with open(file, "r", encoding="utf-8") as f:
            data = safe_load(f)

        if data is None:
            continue

        df = pd.json_normalize(data)

        df["match_id"] = match_id
        match_id += 1

        batch.append(df)

        # Save every 200 matches
        if len(batch) == 200:

            batch_df = pd.concat(batch, ignore_index=True)

            batch_df.to_pickle(
                f"batches/batch_{batch_num:03d}.pkl"
            )

            print(
                f"\nSaved batch {batch_num} "
                f"({len(batch_df)} rows)"
            )

            batch_num += 1
            batch = []

    except Exception as e:
        print(f"\nError in {file}")
        print(e)

# Save last partial batch
if batch:

    batch_df = pd.concat(batch, ignore_index=True)

    batch_df.to_pickle(
        f"batches/batch_{batch_num:03d}.pkl"
    )

    print(f"\nSaved final batch {batch_num}")
```

```python
matches = pickle.load(open('dataset_level1.pkl','rb'))
```

```python
matches.iloc[0]['innings'][0]['1st innings']['deliveries']
```

```python
count = 0
delivery_df = pd.DataFrame()
for index, row in matches.iterrows():
    if count in [75,108,150,180,268,360,443,458,584,748,982,1052,1111,1226,1345]:
        count+=1
        continue
    count+=1
    ball_of_match = []
    batsman = []
    bowler = []
    runs = []
    player_of_dismissed = []
    teams = []
    batting_team = []
    match_id = []
    city = []
    venue = []
    for ball in row['innings'][0]['1st innings']['deliveries']:
        for key in ball.keys():
            match_id.append(count)
            batting_team.append(row['innings'][0]['1st innings']['team'])
            teams.append(row['info.teams'])
            ball_of_match.append(key)
            batsman.append(ball[key]['batsman'])
            bowler.append(ball[key]['bowler'])
            runs.append(ball[key]['runs']['total'])
            city.append(row['info.city'])
            venue.append(row['info.venue'])
            try:
                player_of_dismissed.append(ball[key]['wicket']['player_out'])
            except:
                player_of_dismissed.append('0')
    loop_df = pd.DataFrame({
            'match_id':match_id,
            'teams':teams,
            'batting_team':batting_team,
            'ball':ball_of_match,
            'batsman':batsman,
            'bowler':bowler,
            'runs':runs,
            'player_dismissed':player_of_dismissed,
            'city':city,
            'venue':venue
        })
    delivery_df = pd.concat([delivery_df, loop_df], ignore_index=True)
```

```python
def bowl(row):
    for team in row['teams']:
        if team != row['batting_team']:
            return team
```

```python
delivery_df['bowling_team'] = delivery_df.apply(bowl,axis = 1)
```
